**1**. Elaborar un algoritmo e implementar en python, **el factorial** y **el doble factorial** de un número entero `n` ingresado.
     ### Ejemplo de Ejecución
    ```python

    >>> doblefactorial(3)
        3

    >>> doblefactorial(1987123)
        =???
```


````python
def factorial(numero): 
    if numero < 0:
        raise ValueError("El factorial de un número negativo no existe")

    elif numero == 0: 
        return 1
        
    else: 
        fact = 1
        while(numero > 1): 
            fact *= numero 
            numero -= 1
        return fact 
````

````python
factorial(2)
````

````python
factorial(6)
````

````python
factorial(-2)
````

````python
factorial(5)
````

````python
def doblefactorial(numero): 
    if numero < 0 and not(numero%2): 
        raise ValueError("El doble factorial de un número par negativo no existe")

    elif numero <0 and numero%2:
        fact = 1
        num = numero
        while(num < 1):
          fact *= num
          num +=2
        return numero/fact

    elif numero == 0: 
        return 1
        
    else: 
        fact = 1
        while(numero > 1): 
            fact *= numero 
            numero -= 2
        return fact 
````

**2**. Diseñar un programa que lea dos numero `a` y `b` y diga que tipo de numero son, de a cuerdo a la siguiente clasificación:

**A** [Números defectivos](https://es.wikipedia.org/wiki/Número_defectivo) : la suma de los divisores propios es menor que el número.

**B** [Números abundantes](https://es.wikipedia.org/wiki/Número_abundante): la suma de los divisores es mayor que el número.

**C** [Números amigos](https://es.wikipedia.org/wiki/N%C3%BAmeros_amigos) `a` y `b` tales que a es la suma de los divisores propios de `b` y viceversa.

**D** [Números perfectos](https://es.wikipedia.org/wiki/N%C3%BAmero_perfecto) la suma de todos sus divisores propios, excepto el mismo numero, es igual al número.

**E** [Números primos](https://es.wikipedia.org/wiki/N%C3%BAmero_primo) el número es divisible unicamente por sí mismo y por 1.
 

 ### Ejemplo de Ejecución
``` python

    >>> clase_numeroA(a = 220, b =  284)
        a = 220 y b = 284 son números amigos

    >>> clase_numeroB(a = 24, b = 12 )
        a = 24 es un numero abudante y b = 12 es un número defectivo. 
```


````python
def clase_numero(a,b):
  import numpy as np
  divisores_a = np.array([divisor for divisor in range(1,a+1) if a%divisor == 0 ])
  divisores_b = np.array([divisor for divisor in range(1,b+1) if b%divisor == 0 ])
  if a == np.sum(divisores_b[:-1]) and b == np.sum(divisores_a[:-1]):
    return(print('a =',a,'y b =',b, 'son números amigos'))
  else:
      if len(divisores_a)==2:
          estatus_a = 'primo'
      elif np.sum(divisores_a[:-1]) < a:
          estatus_a = 'defectivo'
      elif np.sum(divisores_a[:-1]) > a:
          estatus_a = 'abundante'
      elif np.sum(divisores_a[:-1]) == a:
          estatus_a = 'perfecto'
      if len(divisores_b)==2:
          estatus_b = 'primo'
      elif np.sum(divisores_b[:-1]) < b:
          estatus_b = 'defectivo'
      elif np.sum(divisores_b[:-1]) > b:
          estatus_b = 'abundante'
      elif np.sum(divisores_b[:-1]) == b:
          estatus_b = 'perfecto'
      return (print('a =', a,'es un número ',estatus_a,'y b =',b,' es un número ',estatus_b))

  
     
````

````python
clase_numero(220,284)
````

````python
clase_numero(a=24,b=8)
````

````python
clase_numero(a=6, b =2)
````

**3**. Determine si un número `n` entero ingresado por el usuario es un [palíndromo](https://en.wikipedia.org/wiki/Palindromic_number), (Retorne `True` en caso afirmativo y `False` en caso contrario) 

```python


    >>> palindromo(3333333)
        True

    >>> palindromo(232)
        True
        
    >>> palindromo(1111349111111)
        False
```




````python
def palindromo(n): # assume n is an integer
    string = str(n)
    if string[::-1] == string:
      r = True
    else:
      r = False
    return r 
````

````python
palindromo(3333333)
````

````python
palindromo(232)
````

````python
palindromo(1111349111111)
````


**4**. Elabore un método para calcular los `x` numeros primos que existen en un intervalos dado por el usuario.


### Ejemplo de Ejecución
```python

>>> primos(5,15):
[5,7,11,13]

>>>primos(1,20):
[1,2,3,5,7,11,13,17,19]

```


````python
def divisores(num):
  import numpy as np
  if num ==0:
    raise ValueError('El número 0 no tiene la división definida')
  else:
    return(np.array([divisor for divisor in range(1,num+1) if num%divisor == 0 ]))
````

````python
divisores(10)
````

````python
def primos(a,b):
  import numpy as np
  primos = []
  for i in range(a,b+1):
    if len(divisores(i)) ==2:
      primos.append(i)
  return primos
  
````

````python
primos(1,20)
````

````python
primos(5,15)
````

````python
v = [22, 32, 42, 12, 22, 31, 41, 11, 12, 232, 24, 12, 22]
````

````python
def quicksort(array):
  import numpy as np
  if len(array)>1:
    pivote = array[np.random.randint(len(array))]
    menores = ([item for item in array if item < pivote])
    iguales = ([item for item in array if item == pivote])
    mayores = ([item for item in array if item > pivote])
    return quicksort(menores)+iguales+quicksort(mayores)
  else:
    return array
````
